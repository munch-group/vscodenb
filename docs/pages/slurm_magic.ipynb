The slurm cell magic allows you to run code in a SLURM job directly from a Jupyter notebook. For example:

```python
%%slurm

s = "Hello from slurm magic!"
```

Variable values updated in slurm job are automatically transferred back to the notebook:

```python
print(s)
```

Specify memory per cpu, number of cpus, wall time, account, etc.

```python
%%slurm -m 8G -c 4 -t 00:10:00 -A xy-drive

s = "Hello again from slurm magic!"
print(s)
```
