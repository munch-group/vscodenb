## Less intrusive progress bars 

````python
from vscodenb import pqdm, prange

for i in prange(10, persist=True):
    time.sleep(0.1)
````

````python
for i in pqdm([1, 2, 3], persist=True):
    time.sleep(0.1)    
````

See a full demo at [here](progress.ipynb).
